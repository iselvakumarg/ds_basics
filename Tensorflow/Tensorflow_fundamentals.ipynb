## Random tensors

Seed function is used to save the state of a random function, so that it can generate same random numbers on multiple executions of the code on the same machine or on different machines (for a specific seed value). The seed value is the previous value number generated by the generator.

```python
# Generate random tensors

random_1 = tf.random.Generator.from_seed(42) # seed is for reproducability
random_1 = random_1.normal(shape=(3, 2))

random_2 = tf.random.Generator.from_seed(42) 
random_2 = random_2.normal(shape=(3, 2))
random_1, random_2, random_1 == random_2
```

```python
#Shuffling the data => when you want to shuffle the data so the inherent order doesn't effect learning

not_shuffled = tf.constant([[10, 7],
                            [3, 4],
                            [1, 2],
                            [5, 9]])

tf.random.shuffle(not_shuffled)
```

```python
# shuffle and manage the same sensor using tensor set_seed

tf.random.set_seed(42) # global level random seed
tf.random.shuffle(not_shuffled, seed=42) # operation level random seed
```

# Matrix Multiplication

There are two rules in tensors(or matrices) need to fulfil the matrix multiplication:
  1. The Inner dimensions must match
  2. The resulting matrix has the shape of the outer dimension

```python
tf.matmul(tensor, tensor)
```

```python
# Matrix Multiplication with python operator "@"

tensor @ tensor
```

```python
# Creating 2 (3, 2) tensors

X = tf.constant([[1, 2],
                 [3, 4],
                 [5, 6]])
Y = tf. constant([[7, 8],
                  [9, 10],
                  [11, 12]])
X, Y
```

```python
# using this two directly leads to dim imcompactible so reshaping Y
X.shape, tf.reshape(Y, shape=(2, 3)).shape
# now the inner dimension are matching
```

```python
X @ tf.reshape(Y, shape=(2, 3))
```

```python
tf.matmul(X, tf.reshape(Y, shape=(2, 3)))
```

```python
# Try reshaping X instead of Y

tf.reshape(X, shape=(2, 3))
```

```python
tf.reshape(X, shape=(2, 3)) @ Y
```

```python
tf.reshape(X, shape=(2, 3)).shape , Y.shape
```

```python
# using Transpose

X, tf.transpose(X), tf.reshape(X, shape=(2, 3))
```

```python
# Matrix Multiplication with Transpose X
tf.matmul(tf.transpose(X), Y)
```

**The Dot Product**

Matrix multiplication is also referred to as the dot product.

The matrix multiplication can be performed using:

* `tf.matmul()`
* `tf.tensordot()`

```python
# dot product on X and Y (either on should be transposed)

tf.tensordot(tf.transpose(X), Y, axes=1)
```

```python
# values of Y, reshape Y and transpose Y

print("Normal Y: ")
print(Y, "\n")

print("Y reshaped to (2, 3): ")
print(tf.reshape(Y, (2, 3)), "\n")

print("Y transposed: ")
print(tf.transpose(Y))
```

Generally, when performing matrix multiplication on two tensors and one of the axes doesn't line up, you will need to transpose (rather than reshape) of the tensors to get satisfy the matrix multiplication rules.

## Aggregating Tensors

Aggregating tensors = condensing them from multiple values down to a small amount of values.

```python
D = tf.constant([-7, -10])
D
```

```python
# Get the absolute values
tf.abs(D)
```

Let's go through the following forms of aggregation:

* Get the minimum
* Get the maximum
* Get the mean of a tensor
* Get the sum of a tensor

```python
# Create a random tensor with values between 0 and 100 of size 50
E = tf.constant(np.random.randint(0, 100, size=50))
E
```

```python
tf.size(E), E.shape, E.ndim
```

```python
# Find the minimum
tf.reduce_min(E)
```

```python
# Find the maximum
tf.reduce_max(E)
```

```python
# Find the mean
tf.reduce_mean(E)
```

```python
# Find the sum
tf.reduce_sum(E)
```

```python
# using tensorflow probability for variance and std

import tensorflow_probability as tfp
tfp.stats.variance(E)
```

```python
tf.math.reduce_std(tf.cast(E, dtype=tf.float32))
```

```python
# Finding variance without importing tfp
tf.math.reduce_variance(tf.cast(E, dtype=tf.float32))
```
